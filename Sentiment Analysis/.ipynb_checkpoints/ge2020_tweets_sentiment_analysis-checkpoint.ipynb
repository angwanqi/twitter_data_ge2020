## Cleaning and Preprocessing the Tweets
Since we are getting the raw dataset from Twitter itself, we will need to do some cleaning and preprocessing on the dataset to prepare it. These are the steps we will take to clean them:
1. Filter out unrelated tweets
2. Remove User Mentions
3. Remove Hashtags
4. Remove URLs
5. Remove New Lines Between Text

First, let's read in the csv file containing the tweets

```python
# Import pandas library
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
pd.set_option('display.max_colwidth', -1)
```

```python
# Read in the csv file and store it in df
df = pd.read_csv('ge2020_tweets.csv')
df.head()
print('Size of dataset is:', df.shape)
```

### Step 1: Filter Out Unrelated Tweets
As our focus is on Singapore's GE2020, we need to filter out the tweets that are unrelated. These tweets are mostly about Ireland General Election and below contains a list of words that are used to filter out the tweets. It is important to note that these words are not entirely encompassing of the possible words used in tweets related to Ireland's GE. However, this word list should be able to remove majority of the unrelated tweets

```python
def find_unrelated_text(text):
    words = ['ireland', 'irish', 'finegael', 'brexit', 'michealmartin',
             'vincentbrowne', 'ff/fg', 'michael_o_regan', 'donaldtrump',
             'taoiseach', 'micheallehane', 'mexican', 'Tipperary',
             'greenparty', 'boyd', 'barrett','socdems','sinnfeinireland',
             'scotland', 'john major','podujanaparty','alankellylabour',
             'greens', 'pensions', 'savecorkcity', 'allah', 'mcguirk',
             'senator', 'eminence', 'green party', 'fffg', 'votegreen',
             'pfg', 'paddy holohan']
    
    contains_text = any([word in text.lower() for word in words])

    return contains_text

# Find unrelated tweets
df['contains_text'] = df['text'].apply(find_unrelated_text)
# Get indexes of unrelated tweets
index_val = df[df['contains_text'] == True ].index
# Drop unrelated tweets from df
df.drop(index_val , inplace=True)

# Print the new size of dataset
print('The new size of dataset after filtering out the unrelated tweets is:', df.shape)
```

### Step 2: Extract Twitter Usernames Mentioned in Each Tweet and Remove Them Thereafter
We want this useful information from each tweet because we can analyse who are the popular figures during the election period. After extracting the information, we will proceed to remove them from the tweet so that it is irrelevant information for our sentiment analysis

```python
import re
```

```python
# Define function to sieve out @users in a tweet:
def mentioned_users(string):
    usernames = re.findall('@[^\s]+', string)
    return usernames

def remove_mentioned_users(string):
    cleaned_tweet = re.sub(r'@\w+ ?', '', string)
    return cleaned_tweet

# Create a new column and apply the functions on the column 'text'
df['mentioned_users'] = df['text'].apply(lambda x: mentioned_users(x))
df['cleaned_text'] = df['text'].apply(lambda x: remove_mentioned_users(x))
```

### Step 3: Extract Twitter Hashtags Mentioned in Each Tweet and Remove Them Thereafter
We want this useful information from each tweet because we can analyse what are the hashtags used during the election period. After extracting the information, we will proceed to remove them from the tweet so that it is irrelevant information for our sentiment analysis

```python
# Define function to sieve out #hashtags in a tweet:
def get_hashtags(string):
    hashtags = re.findall('#[^\s]+', string)
    return hashtags

def remove_hashtags(string):
    cleaned_tweet = re.sub(r'#([^\s]+)', '', string)
    return cleaned_tweet

# Create a new column and apply the functions on the column 'text'
df['hashtags'] = df['text'].apply(lambda x: get_hashtags(x))
df['cleaned_text'] = df['cleaned_text'].apply(lambda x: remove_hashtags(x))
```

### Step 4: Remove URLs

```python
df['cleaned_text'] = df['cleaned_text'].apply(lambda x: re.sub(r'http\S+', '', x))
```

### Step 5: Remove New Lines Between Text

```python
df['cleaned_text'] = df['cleaned_text'].apply(lambda x: x.replace('\n', ''))
```

## Sentiment Analysis of the Tweets
### No. of Positive Sentiments vs No. of Negative Sentiments

Next, we look at what is the distribution of positive and negative tweets. Based on the SentimentIntensityAnalyzer from the NLTK Vader-Lexicon library, this analyser examines the sentiment of a sentence, on how positive, neutral or negative it is.

The analyser returns 4 scores for each sentence namely, positive, negative, neutral and compound. The score compound returns the overall sentiment of a sentence with range of [-1, 1]. For our current purpose, we classify each tweet into 5 classes using the compound score and assign a range of values for each of the classes:

1. Very positive 5  [0.55, 1.00]
2. Positive 4  [0.10, 0.55)
3. Neutral 3  (-0.10, 0.10)
4. Negative 2  (-0.55, -0.10]
5. Very negative 1  [-1.00, -0.55]

As it turns out, analysing the sentiment of the tweets using a rule-based approach is extremely inaccurate because of the nature of the elections. The sentiment of each tweet can be about a particular political party or some could even mention multiple political party. Let's try to get some sentiment analysis done on our tweets!

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
def get_tweet_sentiment(tweet):
    analyzer = SentimentIntensityAnalyzer()
    sentiment_dict = analyzer.polarity_scores(tweet)
    return sentiment_dict
```

```python
df['sentiment_dict'] = df['cleaned_text'].map(get_tweet_sentiment)
df['sentiment_dict'].head()
```

```python
# Get the individual component from sentiment_dict
df['pos'] = df['sentiment_dict'].apply(lambda x: x['pos'])
df['neg'] = df['sentiment_dict'].apply(lambda x: x['neg'])
df['neu'] = df['sentiment_dict'].apply(lambda x: x['neu'])
df['compound'] = df['sentiment_dict'].apply(lambda x: x['compound'])
```

```python
# Assign a class score according to the compound score
def get_sentiment_class(score):
    if (score <= 1.0 and score >= 0.55):
        return 5
    elif (score < 0.55 and score >= 0.10):
        return 4
    elif (score < 0.10 and score > -0.10):
        return 3
    elif (score <= -0.10 and score > -0.55):
        return 2
    elif (score <= -0.55 and score >= -1.00):
        return 1

df['sentiment_class'] = df['compound'].map(get_sentiment_class)
```

```python
# Verify if the classification assignment is correct:
df[['compound', 'sentiment_class']].head()
```

### Next, we can make a seaborn countplot to show us the distribution of sentiment classes in the dataset:
