#### Step 1: Filter Unrelated Tweets
As our focus is on Singapore's GE2020, we need to filter out the tweets that are unrelated. These tweets are mostly about Ireland General Election and below contains a list of words that are used to filter out the tweets. It is important to note that these words are not entirely encompassing of the possible words used in tweets related to Ireland's GE. However, this word list should be able to remove majority of the unrelated tweets

```python
def find_unrelated_text(text):
    words = ['ireland', 'irish', 'finegael', 'brexit', 'michealmartin',
             'vincentbrowne', 'ff/fg', 'michael_o_regan', 'donaldtrump',
             'taoiseach', 'micheallehane', 'mexican', 'Tipperary',
             'greenparty', 'boyd', 'barrett','socdems','sinnfeinireland',
             'scotland', 'john major','podujanaparty','alankellylabour',
             'greens', 'pensions', 'savecorkcity', 'allah', 'mcguirk',
             'senator', 'eminence', 'green party', 'fffg', 'votegreen',
             'pfg', 'paddy holohan']
    
    contains_text = any([word in text.lower() for word in words])

    return contains_text

# Find unrelated tweets
df['contains_text'] = df['text'].apply(find_unrelated_text)
# Get indexes of unrelated tweets
index_val = df[df['contains_text'] == True ].index
# Drop unrelated tweets from df
df.drop(index_val , inplace=True)
```

1. Step 2: Remove user mentions (e.g. '@user') from all tweets
2. Step 3 : Remove URLs
3. Step 4: Remove Hashtag Symbol
4. Step 5: Remove New Lines between text

```python
# remove twitter handles (@user)
df['clean_tweet'] = df['text'].apply(lambda x: re.sub(r'@\w+ ?', '', x))
df['clean_tweet'] = df['clean_tweet'].apply(lambda x: re.sub(r'http\S+', '', x))
df['clean_tweet'] = df['clean_tweet'].apply(lambda x: re.sub('#',  '', x))
df['clean_tweet'] = df['clean_tweet'].apply(lambda x: x.replace("\n", ""))
df[['text','clean_tweet']].head(10)
```

# Visualising tweets using Word Cloud
To start off, we will look at tweets from all 13 days and create a word cloud to find what are the most common words used in all tweets across all days. First, we will have to join all tweets together to create one long ultra string. 

```python
text = " ".join(tweet.lower() for tweet in df.clean_tweet)
print ("There are {} words in the combination of all tweets.".format(len(text)))
```

Next, we will use the stopwords list from word cloud to remove stopwords from our long string. I have also added some of my stopwords.

```python
# Create stopword list:
stopwords = set(STOPWORDS)
stopwords.update(['ge2020', 'singapore', 'party', 'political','will','vote', 'amp','one',
                  'sgelection','people', 'singaporevote','election','singaporevotes'])

# Generate a word cloud image
wordcloud = WordCloud(width=800, height=500, random_state=21,stopwords=stopwords, 
                      background_color="white").generate(text)

# Display the generated image:
# the matplotlib way:
plt.figure(figsize=(10, 7))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis('off')
plt.title("Most Common words in Tweets")
plt.show()
```

Now, let's generate a word cloud for each day to find out what's the common words used on each day!

```python
dates = ['2020-06-26', '2020-06-27', '2020-06-28', '2020-06-29', '2020-06-30',
     '2020-07-01', '2020-07-02', '2020-07-03', '2020-07-04', '2020-07-05',
     '2020-07-06', '2020-07-07', '2020-07-08', '2020-07-09', '2020-07-10',
     '2020-07-11', '2020-07-12', '2020-07-13', '2020-07-14', '2020-07-15']

text_by_date = []
total_words = 0
for date in dates:
    current_date_df = df.loc[df['date'] == date]
    text = " ".join(tweet.lower() for tweet in current_date_df.clean_tweet)
    text_by_date.append(text)
```

```python
fig = plt.figure(figsize=(30, 20))
i = 0
for text, date in zip(text_by_date, dates):
    # Generate a word cloud image
    wordcloud = WordCloud(width=1000, height=500, random_state=21,stopwords=stopwords, 
                          background_color="white").generate(text)
    
    ax = fig.add_subplot(5,4,i+1)
    ax.imshow(wordcloud, interpolation="bilinear")
    ax.axis('off')
    i = i + 1
    ax.set_title("Most Common words in Tweets on " + date)
```

```python
def get_tweet_sentiment(tweet):
    analyzer = SentimentIntensityAnalyzer()
    sentiment_dict = analyzer.polarity_scores(tweet)
    return sentiment_dict
```

```python
all_tweets_df = df[['clean_tweet','date']]
all_tweets_df.head()
```

```python
all_tweets_df['sentiment'] = all_tweets_df['clean_tweet'].map(get_tweet_sentiment)
all_tweets_df['sentiment'].head()
```

```python
all_tweets_df['pos'] = all_tweets_df['sentiment'].apply(lambda x: x['pos'])
all_tweets_df['neg'] = all_tweets_df['sentiment'].apply(lambda x: x['neg'])
all_tweets_df['neu'] = all_tweets_df['sentiment'].apply(lambda x: x['neu'])
all_tweets_df['compound'] = all_tweets_df['sentiment'].apply(lambda x: x['compound'])
```

```python
# Very positive ‘5’ — [0.55, 1.00]
# Positive ‘4’ — [0.10, 0.55)
# Neutral ‘3’ — (-0.10, 0.10)
# Negative ‘2’ — (-0.55, -0.10]
# Very negative ‘1’ — [-1.00, -0.55]

def get_sentiment_class(score):
    if (score <= 1.0 and score >= 0.55):
        return 5
    elif (score < 0.55 and score >= 0.10):
        return 4
    elif (score < 0.10 and score > -0.10):
        return 3
    elif (score <= -0.10 and score > -0.55):
        return 2
    elif (score <= -0.55 and score >= -1.00):
        return 1
```

```python
all_tweets_df['sentiment_class'] = all_tweets_df['compound'].map(get_sentiment_class)
```

```python
import seaborn as sns

# Distribution of sentiment_class
plt.figure(figsize = (10,5))
sns.set_palette('PuBuGn_d')
sns.countplot(all_tweets_df['sentiment_class'])
plt.title('Countplot of sentiment_class')
plt.xlabel('sentiment_class')
plt.ylabel('No. of classes')
plt.show()
```
